```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('titanic-Dataset.csv')
```

```python

df.drop_duplicates(inplace=True)
print(df.isnull().sum())
```

```python
df = df.fillna(0)
```

```python
print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=20, color='blue') 
plt.title("Distribution of Ticket Fares")
plt.xlabel("Fare Price")
plt.ylabel("Frequency")
plt.show()
```


1. Majority of Passengers:** Most passengers purchased low-cost tickets, concentrated in the **0-50** range.
2. Wealth Distribution:** There is a significant gap between the general passengers and the few who paid high fares (above 200), indicating a clear distinction between travel classes.
3. Outliers: There are a few extreme values (outliers) on the right side of the graph representing luxury suites.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['Fare'] = df['Fare'].fillna(0)

pclass_revenue = df.groupby('Pclass')['Fare'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=pclass_revenue.index.astype(str), y=pclass_revenue.values, hue=pclass_revenue.index.astype(str), palette='viridis', legend=False)

plt.title("Total Revenue by Passenger Class", fontsize=14)
plt.xlabel("Passenger Class (1, 2, 3)", fontsize=12)
plt.ylabel("Total Fare Collected", fontsize=12)

plt.show()
print(pclass_revenue)
```
